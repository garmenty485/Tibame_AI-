# Please use POST to find that how many train from "板橋" to "彰化" at 12:00 today.
url: https://www.thsrc.com.tw/ArticleContent/a3b630bb-1066-4352-a1ef-58c7b4e8ef7c

```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import json

payload = {

"SearchType":"S",
"Lang":"TW",
"StartStation":"NanGang",
"EndStation":"ZuoYing",
"OutWardSearchDate":"2022/12/25",
"OutWardSearchTime":"21:30",
"ReturnSearchDate":"2022/12/25",
"ReturnSearchTime":"21:30",
"DiscountType":""
    
}


url = 'https://www.thsrc.com.tw/TimeTable/Search'
response = requests.post(url, data = payload)
#print(response.text)

result = json.loads(response.text)
#print(result['data']['DepartureTable']['TrainItem'])

for i in result['data']['DepartureTable']['TrainItem']:
    print('Train number: {}, Departure time: {}, Destination time: {}'.format(i['TrainNumber'], i['DepartureTime'], i['DestinationTime']))
```

```python
# iterate over the dictionary
for ii in result['data']['DepartureTable']['TrainItem']:
    output = ''
    for i, (key, value) in enumerate(ii.items()):
        if i != 2:
            output += (str(key)+':'+str(value)+' ,') 
    
    # break the loop after printing the third key-value pair
        if i == 2:
            output += (str(key)+':'+str(value)+' ****')
            print(output)
            break
```
